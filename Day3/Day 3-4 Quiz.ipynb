## Question 12
What is the difference between soft and hard links?

- Soft link is created on a directory where hard link is created on a file
- In a soft link, if the source file is deleted, then the link is broken, but in a hard link, deleting a source file will not impact the linked file
- In a hard link, if the source file is deleted, then the link is broken, but in a soft link, deleting a source file will not impact the linked file
- Soft link is created when the original source file exists where a hard link does not require any source file
